```python
import soundfile as sf
import matplotlib.pyplot as plt

f_name = 'sounddata.h'
f = open(f_name, 'w')

max_len = []

def func(i, max_len):
    names = '1'
    if(i == -2):
        names = 'Start_of_transmission'
    elif(i == -1):
        names = 'Start_of_transmission'
    elif(i == 26):
        names = 'End_of_transmission'
    else:
        names = '' + str(i)

    soundfile = names+'.wav'
    soundfile_path = 'C:/Users/ulgeo/OneDrive/Desktop/Voice_numbers/WAV/'+soundfile

    data_in, datasamplerate = sf.read(soundfile_path)
    # This means stereo so extract one channel 0
    if len(data_in.shape)>1:
        data_in = data_in[:,0]

    import samplerate
    converter = 'sinc_best'  # or 'sinc_fastest', ...
    desired_sample_rate = 11000.0
    ratio = desired_sample_rate/datasamplerate
    data_out = samplerate.resample(data_in, ratio, converter)
    maxValue = max(data_out)
    minValue = min(data_out)
    vrange = (maxValue - minValue) 

    m68code = "/*    File "+soundfile+ "\r\n *    Sample rate "+str(int(desired_sample_rate)) +" Hz\r\n */\r\n"
    
    
    
    if(i == -2):
        m68code += "#define WAV_DATA_LENGTH "+str(len(data_out))+" \r\n"
        m68code += "#define WAV_DATA WAV_DATA_" +names +"\r\n"
    
    elif(names == 'Start_of_transmission'):
        m68code += "#define len_s "+str(len(data_out))+" \r\n"
        m68code += "#define samp_s WAV_DATA_" +names +"\r\n"
    elif(names == 'End_of_transmission'):
        m68code += "#define len_e "+str(len(data_out))+" \r\n"
        m68code += "#define samp_e WAV_DATA_" +names +"\r\n"
    else:
        m68code += "#define len_"+names+" "+str(len(data_out))+" \r\n"
        m68code += "#define samp_" +names +" WAV_DATA_" +names +"\r\n"

    m68code += "uint8_t WAV_DATA_" +names+ "[] = {\r\n    "

    maxitemsperline = 64
    itemsonline = maxitemsperline
    firstvalue = 0
    lastvalue = 0
    for v in data_out:
        # scale v to between 0 and 1
        isin = (v-minValue)/vrange   
        max_len.append(isin)
        v =  int((isin * 255))
        vstr = str(v)
        if (firstvalue==0):
            firstvalue= v
        lastvalue = v
        m68code+=vstr
        itemsonline-=1
        if (itemsonline>0):
            m68code+=','
        else:
            itemsonline = maxitemsperline
            m68code+=',\n    '

    # keep track of first and last values to avoid
    # blip when the loop restarts.. make the end value
    # the average of the first and last. 
    end_value = int( (firstvalue + lastvalue) / 2)
    m68code+=str(end_value)+'    \r\n}; //'+ names +'\r\n'
    f.write(m68code)

for i in range(-1,27):
    func(i, max_len)
    
f.close()

print(max(max_len))
```
